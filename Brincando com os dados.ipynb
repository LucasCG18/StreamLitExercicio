```python
# quero transformar a coluna diagnosis em boolean
# M = maligno (1)  B = benigno (0)
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
df['diagnosis'].value_counts()
```

```python
# mudar para boolean
df['diagnosis'] = df['diagnosis'].astype(bool)
print(df['diagnosis'].dtype)
```

```python
# Criando uma figura com dois subplots (um ao lado do outro)
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Histograma para a varivel 'raio'
sns.histplot(df['radius_mean'], bins=30, kde=True, ax=axes[0], color='salmon')
axes[0].set_title('Distribuio do Raio Mdio')
axes[0].set_xlabel('Raio Mdio')
axes[0].set_ylabel('Frequncia')
# Boxplot para a varivel 'textura'
sns.boxplot(x=df['texture_mean'], ax=axes[1], color='lightblue')
axes[1].set_title('Boxplot da Textura Mdia')
axes[1].set_xlabel('Textura Mdia')
axes[1].set_ylabel('Valor')

# Ajusta o layout para no haver sobreposio
plt.tight_layout()
plt.show()
```

```python
#grafico de spearman do raio em relaxao ao diagnostico
plt.figure(figsize=(10, 6))
sns.boxplot(x='diagnosis', y='radius_mean', data=df, palette='Set2')
plt.title('Boxplot do Raio Mdio por Diagnstico')
plt.xlabel('Diagnstico (0 = Benigno, 1 = Maligno)')
plt.ylabel('Raio Mdio')
plt.show()

# grafico de spearman da textura em relaxao ao diagnostico
plt.figure(figsize=(10, 6))
sns.boxplot(x='diagnosis', y='texture_mean', data=df, palette='Set3')
plt.title('Boxplot da Textura Mdia por Diagnstico')
plt.xlabel('Diagnstico (0 = Benigno, 1 = Maligno)')
plt.ylabel('Textura Mdia')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()

# Aplicando StandardScaler
numerical_cols = df.select_dtypes(include=['float32', 'float64']).columns.tolist()
#df_standard_scaled = df[numerical_cols].copy() --- IGNORE ---
#df_standard_scaled[numerical_cols] = standard_scaler.fit_transform(df[numerical_cols]) --- IGNORE ---

# Aplicando MinMaxScaler
df_minmax_scaled = df[numerical_cols].copy()
df_minmax_scaled[numerical_cols] = minmax_scaler.fit_transform(df[numerical_cols])
```

```python
# Cria boxplots para cada grupo de 3 variveis
for i in range(0, len(numerical_cols), 3):
    cols_group = numerical_cols[i:i+3]
    sns.boxplot(data=df_minmax_scaled[cols_group])
    plt.title(f'Distribuio das Variveis: {", ".join(cols_group)}')
    plt.ylabel('Valor')
    plt.xlabel('Variveis')
    plt.show()
```

```python
#grafico de dispersao de textura com raio
plt.figure(figsize=(10, 6))
sns.scatterplot(x='radius_mean', y='texture_mean', hue='diagnosis', data=df, alpha=0.6, palette='Set1')
plt.title('Relao entre Raio Mdio e Textura Mdia')
plt.xlabel('Raio Mdio')
plt.ylabel('Textura Mdia')
plt.legend(title='Diagnstico', labels=['Benigno (0)', 'Maligno (1)'])
plt.show()
```

```python
# Grfico de barras para benigno e maligno com raio maior que 15
df_maior_15 = df[df['radius_mean'] > 15]
df_menor_15 = df[df['radius_mean'] <= 15]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Raio maior que 15
df_maior_15['diagnosis'].value_counts().plot(kind='bar', ax=axes[0], color=['skyblue', 'salmon'])
axes[0].set_title('Diagnstico (Raio > 15)')
axes[0].set_xlabel('Diagnstico')
axes[0].set_ylabel('Contagem')
axes[0].set_xticklabels(['Benigno (0)', 'Maligno (1)'], rotation=0)

# Raio menor ou igual a 15
df_menor_15['diagnosis'].value_counts().plot(kind='bar', ax=axes[1], color=['skyblue', 'salmon'])
axes[1].set_title('Diagnstico (Raio  15)')
axes[1].set_xlabel('Diagnstico')
axes[1].set_ylabel('Contagem')
axes[1].set_xticklabels(['Benigno (0)', 'Maligno (1)'], rotation=0)

plt.tight_layout()
plt.show()
```

```python
# juntar as variaveis para ver a correlacao
corr_matrix = df.corr(method='spearman', numeric_only=True)
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Matriz de Correlao (Spearman)')
plt.show()
```

```python
# juntar as variaveis para ver a correlacao
corr_matrix = df.corr(method='kendall', numeric_only=True)
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Matriz de Correlao (Spearman)')
plt.show()
```

```python
#grafico para ver correlao entre diagnostico e raio worst
plt.figure(figsize=(10, 6))
sns.boxplot(x='diagnosis', y='radius_worst', data=df, palette='Set1')
plt.title('Boxplot do Raio Worst por Diagnstico')
plt.xlabel('Diagnstico (0 = Benigno, 1 = Maligno)')
plt.ylabel('Raio Worst')
plt.show()
```
